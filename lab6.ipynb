## Lab 6 - Bài 1

```python
# Code lab 6 bài 1 ở đây
class ChuNhat:
    def __init__(self, dai, rong):
        self.dai = dai
        self.rong = rong

    def get_chu_vi(self):
        return 2 * (self.dai + self.rong)

    def get_dien_tich(self):
        return self.dai * self.rong

    def xuat(self):
        print("=" * 30)
        print("Thông tin Hình Chữ Nhật:")
        print(f"Chiều dài: {self.dai}")
        print(f"Chiều rộng: {self.rong}")
        print(f"Diện tích: {self.get_dien_tich()}")
        print(f"Chu vi: {self.get_chu_vi()}")
        


class Vuong(ChuNhat):
    def __init__(self, canh):
        super().__init__(canh, canh)

    def xuat(self):
        print("=" * 30)
        print("Thông tin Hình Vuông:")
        print(f"Cạnh: {self.dai}")
        print(f"Diện tích: {self.get_dien_tich()}")
        print(f"Chu vi: {self.get_chu_vi()}")
        
    
    try:
        dai_cn1 = float(input("Nhập chiều dài Hình chữ nhật 1: "))
        rong_cn1 = float(input("Nhập chiều rộng Hình chữ nhật 1: "))
    except ValueError:
        print("Lỗi: Đầu vào không hợp lệ. Vui lòng nhập số.")
        exit()

    try:
        dai_cn2 = float(input("Nhập chiều dài Hình chữ nhật 2: "))
        rong_cn2 = float(input("Nhập chiều rộng Hình chữ nhật 2: "))
    except ValueError:
        print("Lỗi: Đầu vào không hợp lệ. Vui lòng nhập số.")
        exit()

    try:
        canh_vu = float(input("Nhập cạnh Hình vuông: "))
    except ValueError:
        print("Lỗi: Đầu vào không hợp lệ. Vui lòng nhập số.")
        exit()

    cn1 = ChuNhat(dai_cn1, rong_cn1)
    cn2 = ChuNhat(dai_cn2, rong_cn2)
    vu = Vuong(canh_vu)

    print("\n--- KẾT QUẢ ---")
    cn1.xuat() 
    cn2.xuat() 
    vu.xuat()  
```

## Lab 6 - Bài 3

```python
# Code lab 6  bài 3 ở đây
import math
class SinhVienPoly:   
    def __init__(self, ho_ten, nganh):
        self.ho_ten = ho_ten
        self.nganh = nganh

    def get_diem(self):
        pass 

    def get_hoc_luc(self):
        diem = self.get_diem()
        
        if diem is None or (isinstance(diem, (int, float)) and diem < 0):
            return "Chưa có điểm hợp lệ"
        
        if diem < 5:
            return "Yếu"
        elif diem < 7:  
            return "Trung bình"
        elif diem < 8: 
            return "Khá"
        elif diem < 9: 
            return "Giỏi"
        else:
            return "Xuất sắc"

    def xuat(self):
        diem = self.get_diem()
        hoc_luc = self.get_hoc_luc()
        
        print("-" * 45)
        print(f"Họ tên: {self.ho_ten}")
        print(f"Ngành: {self.nganh}")
        
        diem_hien_thi = f"{diem:.2f}" if isinstance(diem, (int, float)) else "Chưa tính được"
        print(f"Điểm trung bình: {diem_hien_thi}")
        print(f"Học lực: {hoc_luc}")
        print("-" * 45)
        
class SinhVienIT(SinhVienPoly):
    def __init__(self, ho_ten, java, html, css):
        super().__init__(ho_ten, "Công nghệ thông tin")
        self.java = java
        self.html = html
        self.css = css

    def get_diem(self):
        return (2 * self.java + self.html + self.css) / 4

class SinhVienBiz(SinhVienPoly):
    def __init__(self, ho_ten, marketing, sales):
        super().__init__(ho_ten, "Kinh doanh")
        self.marketing = marketing
        self.sales = sales

    def get_diem(self):
        return (2 * self.marketing + self.sales) / 3
    
    print("--- THÔNG TIN SINH VIÊN IT ---")
    sv_it = SinhVienIT("Nguyễn Thị Thảo", java=7, html=8, css=9)
    sv_it.xuat()

    print("\n--- THÔNG TIN SINH VIÊN BIZ 1 ---")
    sv_biz = SinhVienBiz("Trần Anh Tú", marketing=9, sales=8)
    sv_biz.xuat()
    
    print("\n--- THÔNG TIN SINH VIÊN BIZ 2 ---")
    sv_biz_2 = SinhVienBiz("Lê Phúc Vinh", marketing=4, sales=6)
    sv_biz_2.xuat()
```

## Lab 6 - Bài 4

```python
# code lab 6 bài 4 ở đây
import math
class SinhVienPoly:
    def __init__(self, ho_ten, nganh):
        self.ho_ten = ho_ten
        self.nganh = nganh

    def get_diem(self):
        pass 

    def get_hoc_luc(self):
        diem = self.get_diem()
        
        if not isinstance(diem, (int, float)) or diem < 0:
            return "Chưa có điểm hợp lệ"
        
        if diem < 5:
            return "Yếu"
        elif diem < 7:
            return "Trung bình"
        elif diem < 8:
            return "Khá"
        elif diem < 9:
            return "Giỏi"
        else:
            return "Xuất sắc"

    def xuat(self):
        diem = self.get_diem()
        hoc_luc = self.get_hoc_luc()
        
        print("-" * 45)
        print(f"Họ tên: {self.ho_ten}")
        print(f"Ngành: {self.nganh}")
        
        diem_hien_thi = f"{diem:.2f}" if isinstance(diem, (int, float)) else "Chưa tính được"
        print(f"Điểm trung bình: {diem_hien_thi}")
        print(f"Học lực: {hoc_luc}")
        print("-" * 45)

class SinhVienIT(SinhVienPoly):
    def __init__(self, ho_ten, java, html, css):
        super().__init__(ho_ten, "Công nghệ thông tin")
        self.java = java
        self.html = html
        self.css = css

    def get_diem(self):
        return (2 * self.java + self.html + self.css) / 4

class SinhVienBiz(SinhVienPoly):
    def __init__(self, ho_ten, marketing, sales):
        super().__init__(ho_ten, "Kinh doanh")
        self.marketing = marketing
        self.sales = sales

    def get_diem(self):
        return (2 * self.marketing + self.sales) / 3
    
danh_sach_sv = []

def nhap_danh_sach_sinh_vien():
    while True:
        try:
            print("\n--- NHẬP SINH VIÊN ---")
            ho_ten = input("Nhập Họ và Tên: ")
            
            print("Chọn Ngành học:")
            print("1. Công nghệ thông tin (IT)")
            print("2. Kinh doanh (Biz)")
            chon_nganh = input("Nhập lựa chọn (1/2): ")
            
            if chon_nganh == '1':
                java = float(input("Nhập điểm Java: "))
                html = float(input("Nhập điểm HTML: "))
                css = float(input("Nhập điểm CSS: "))
                sv = SinhVienIT(ho_ten, java, html, css)
                danh_sach_sv.append(sv)
                print(f"Đã thêm sinh viên IT: {ho_ten}")

            elif chon_nganh == '2':
                marketing = float(input("Nhập điểm Marketing: "))
                sales = float(input("Nhập điểm Sales: "))
                sv = SinhVienBiz(ho_ten, marketing, sales)
                danh_sach_sv.append(sv)
                print(f"Đã thêm sinh viên Biz: {ho_ten}")
            
            else:
                print("Lựa chọn không hợp lệ.")
                continue

        except ValueError:
            print("Lỗi: Đầu vào điểm phải là số. Vui lòng thử lại.")
            continue

        tiep_tuc = input("Bạn có muốn nhập tiếp không? (y/n): ")
        if tiep_tuc.lower() != 'y':
            break

def xuat_thong_tin_danh_sach():
    if not danh_sach_sv:
        print("\nDanh sách sinh viên rỗng.")
        return
        
    print("\n--- DANH SÁCH SINH VIÊN ---")
    for i, sv in enumerate(danh_sach_sv):
        print(f"STT: {i + 1}")
        sv.xuat()

def xuat_danh_sach_hoc_luc_gioi():
    if not danh_sach_sv:
        print("\nDanh sách sinh viên rỗng.")
        return
        
    ds_gioi = [sv for sv in danh_sach_sv if sv.get_hoc_luc() == "Giỏi"]
    
    if not ds_gioi:
        print("\nKhông có sinh viên nào đạt học lực Giỏi.")
        return
        
    print("\n--- DANH SÁCH SINH VIÊN HỌC LỰC GIỎI ---")
    for i, sv in enumerate(ds_gioi):
        print(f"STT: {i + 1}")
        sv.xuat()

def sap_xep_danh_sach_theo_diem():
    if not danh_sach_sv:
        print("\nDanh sách sinh viên rỗng.")
        return
    
    try:
        danh_sach_sv.sort(key=lambda sv: sv.get_diem(), reverse=True)
        print("\nĐã sắp xếp danh sách sinh viên theo điểm (từ cao xuống thấp).")
        xuat_thong_tin_danh_sach()
    except TypeError:
        print("\nLỗi: Không thể sắp xếp do có sinh viên chưa có điểm hợp lệ.")

def menu():
    print("\n====== CHƯƠNG TRÌNH QUẢN LÝ SINH VIÊN ======")
    print("1. Nhập danh sách sinh viên")
    print("2. Xuất thông tin danh sách sinh viên")
    print("3. Xuất danh sách sinh viên có học lực Giỏi")
    print("4. Sắp xếp danh sách sinh viên theo điểm")
    print("5. Kết thúc")
    print("===========================================")

def main():
    while True:
        menu()
        lua_chon = input("Vui lòng chọn chức năng: ")
        if lua_chon == '1':
            nhap_danh_sach_sinh_vien()
        elif lua_chon == '2':
            xuat_thong_tin_danh_sach()
        elif lua_chon == '3':
            xuat_danh_sach_hoc_luc_gioi()
        elif lua_chon == '4':
            sap_xep_danh_sach_theo_diem()
        elif lua_chon == '5':
            print("Đã kết thúc chương trình.")
            break
        else:
            print("Lựa chọn không hợp lệ. Vui lòng chọn lại.")
if __name__ == "__main__":
    main()
```
